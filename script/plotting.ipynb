# Example plotting from the docs

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import mesa_reader to make its classes accessible
import mesa_reader as mr
```

## star_plus_point_mass

```python
# make a MesaData object from a history file

# NOTE: Edit the argument to your location of the LOGS dir.
logs_dir = '/home2/yousousen/software_etc/mesa/mesa-r22.11.1/binary/working_with_mesa/star_plus_point_mass/LOGS1/'

h = mr.MesaData(logs_dir + 'history.data') # Plotting history data.

x = h.model_number
y = h.star_mass
plt.plot(x, y)
```
